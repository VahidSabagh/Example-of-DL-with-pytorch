```python
import numpy as np
import matplotlib.pyplot as plt
```

Linear Regression

```python
x=np.random.rand(100,1)
y=1+2*x + (0.1 * np.random.randn(100,1))  #noise: np.random.randn(100,1) 
```

## Separate training and test data
###### From Shuffle to randomly select samples

```python
idx = np.arange(100)
np.random.shuffle(idx)
```

```python
train_idx = idx[:80]
test_idx = idx[80:]

x_train, y_train = x[train_idx],y[train_idx]
x_test, y_test = x[test_idx],y[test_idx]
```

```python
plt.plot(x_train,y_train,'o')
```

```python
plt.plot(x_test,y_test,'o')
```

# Model

```python
a = np.random.randn(1)
b = np.random.randn(1)
print('a,b:  ', a, b)
yp = a + b * x_train
```

# Loss

```python
error = (y_train - yp)
loss = (error ** 2).mean()
loss #MSE
```

# Gradient calculations

```python
a_grad = -2 * error.mean()
b_grad = -2 * (x_train * error).mean()
print(a_grad, b_grad)
```

# update

```python
lr = 0.01        #learning rate
```

```python
a = a - lr * a_grad
b = b - lr * b_grad

print('a,b:  ', a, b)
```

# Main

```python
np.random.seed(2)
a = np.random.randn(1)
b = np.random.randn(1)
lr = 0.1
n_epochs = 100
print('a,b:  ', a, b)

for epoch in range(n_epochs):
    # model
    yp = a + b * x_train

    # loss
    error = (y_train - yp)
    loss = (error ** 2).mean()

    # gradiant
    a_grad = -2 * error.mean()
    b_grad = -2 * (x_train * error).mean()

    # update
    a = a - lr * a_grad
    b = b - lr * b_grad
    
    print(loss)
print('a,b  --->  ', a, b)
```

```python
plt.plot(x_test, y_test, "ro")

yhat = a + b * x_test

plt.plot(x_test, yhat, 'g')
```

# torch1

```python
import torch

y_train = torch.from_numpy(y_train)
x_train = torch.from_numpy(x_train)    #Numpy array is converted to tensor
```

```python
# np.random.seed(2)
# a = np.random.randn(1)
# b = np.random.randn(1)

torch.manual_seed(10)
a = torch.randn(1)
b = torch.randn(1)


lr = 0.1 
n_epochs = 1000
print('a,b:  ', a, b)

for epoch in range(n_epochs):
    # model
    yp = a + b * x_train

    # loss
    error = (y_train - yp)
    loss = (error ** 2).mean()

    # gradiant
    a_grad = -2 * error.mean()
    b_grad = -2 * (x_train * error).mean()

    # update
    a = a - lr * a_grad
    b = b - lr * b_grad
    
    print(loss)
print('a,b  --->  ', a, b)
```
